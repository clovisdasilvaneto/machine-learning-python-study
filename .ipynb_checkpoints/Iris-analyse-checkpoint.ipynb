```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = np.genfromtxt('../datasets/iris.data', delimiter=',', usecols=[0,1,2,3]);

setosa = iris[:50];
versicolor = iris[50:100];
virginica = iris[100:150];
```

```python
plt.plot(setosa[:,0], c='Red', marker='o', ls='-', label="Comp. Sépala Iris Setosa")
plt.plot(versicolor[:,0], c='Blue', marker='o', ls='-', label="Comp. Sépala Iris Versicolor")
plt.plot(virginica[:,0], c='Green', marker='o', ls='-', label="Comp. Sépala Iris Virginica")
plt.
plt.legend()
plt.show()
```
