```python
import matplotlib.pyplot as plt;
import numpy as np;
import pydataset;

%matplotlib inline
```

```python
salary = np.array([100, 200, 300, 400, 550]);
```

```python
plt.plot(salary)
```

```python
plt.plot(salary, c='Black', ls='--');
```

```python
plt.plot(salary, c='Black', ls='--', marker='s');
```

```python
plt.plot(salary, c='Red', ls='--', marker='s', ms='20');
plt.show()
```

```python
salary2 = [1000, 2000, 3000, 4000, 5000];

plt.plot(salary2, c='Red', ms=10, ls='-', marker='s', label='Salary 2');
plt.plot(salary, c='Blue', ms=10, ls='-', marker='^', label='Salary 1');
plt.legend();
plt.show();
```

```python
salary2 = [1000, 2000, 3000, 4000, 5000];

plt.plot(salary2, c='Red', ms=10, ls='-', marker='s', label='Salary 2');
plt.plot(salary, c='Blue', ms=10, ls='-', marker='^', label='Salary 1');
plt.legend(loc='upper right');
plt.show();
```

```python
iris = pydataset.data('iris');

setosa = iris[iris['Species'] == 'setosa'];
virginica = iris[iris['Species'] == 'virginica'];
versicolor = iris[iris['Species'] == 'versicolor'];

versicolor.head()
```

```python


plt.scatter(versicolor['Petal.Length'],versicolor['Sepal.Width'],c='red', alpha=0.4)
plt.scatter(virginica['Petal.Length'],virginica['Sepal.Width'],c='green', alpha=0.4)
plt.scatter(setosa['Petal.Length'],setosa['Sepal.Width'],c='blue', alpha=0.4)
```
