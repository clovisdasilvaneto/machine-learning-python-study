```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris;

iris = load_iris();
print(iris.data)
```

```python
print(iris.target)
```

```python
## Esses pontos abaixos representam cada ponto identificado pelo iris.data
print(iris.feature_names)
```

```python
def getIrisName(index): 
    return iris.target_names[index];

print(iris.target_names)
```

```python
x = iris.data;
y = iris.target;

print(x,y)
```

```python
print(y.shape, x.shape)
```

### I'm going to use Kneighbors

Irei usar o algoritimo de k-neighbors para poder obter a predição das datas mais comuns.

```python
from sklearn.neighbors import KNeighborsClassifier

## Onde neightbors é o numero de visinhos mais comum que eu quero que meu modelo se baseie para aprender
knn = KNeighborsClassifier(n_neighbors = 1);
```

```python
print(knn)
```

Agora irei treinar meu modelo de acordo com a data que eu tenho =D

```python
## I'm going to training my model, using de fit() function
knn.fit(x,y)
```

Agora iremos criar predições com nosso model já treinado ehehehhe

```python
## com knn.predict([array]), eu estou predisindo meu model de acordo com a data encontrada no modelo
prediction = getIrisName(knn.predict([3,5,4,2]));

print('o tipo da flor é: ', prediction[0])
```
