## Teste de gráfico de dispersão

```python
from matplotlib import pyplot as plt

number_of_friends = [300, 300, 370, 200, 340, 300, 240, 350];
minutes_on_line = [360, 280, 300, 230, 320, 223, 223, 256];
friends_labels = ['jenny', 'clovis', 'davi', 'nevinha', 'filha de nevinha', 'ana', 'monica', 'jorge'];

plt.scatter(number_of_friends, minutes_on_line);

for label, friends, minutes in zip(friends_labels, number_of_friends, minutes_on_line):
    plt.annotate(label,
            xy=(friends,minutes),
            xytext=(5,-5),
            textcoords='offset points');  
    
plt.title("Minutos diários vs número de amigos");
plt.xlabel('# de amigos');
plt.ylabel('minutos diários passados no site');

plt.show();
```
