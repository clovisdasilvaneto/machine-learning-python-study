## Teste de gr√°fico em linhas

```python
from matplotlib import pyplot as plt

variance = [1,2,4,8,16,32,64,128,256];
bias_squarred = [256,128,64,32,16,8,4,2,1];
total_error = [x + y for x, y in zip(variance, bias_squarred)];
xs = [i for i,_ in enumerate(variance)];

plt.plot(xs, variance, 'b-', label='variance');
plt.plot(xs, bias_squarred, 'g-', label='bias^2');
plt.plot(xs, total_error, 'r:', label='total error');


#labels
plt.legend(loc=9);
plt.xlabel('complexdade do modelo');
plt.title('Compromisso entre Polariza√ß√£o e Vari√¢ncia');
plt.show()
```
